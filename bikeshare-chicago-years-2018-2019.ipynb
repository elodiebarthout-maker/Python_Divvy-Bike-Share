```python
#Joining 2018 & 2019 datasets
all_trips = pd.concat([data_2018, data_2019], ignore_index = True)
all_trips.describe
```

```python
#Data cleaning
#Checking for inconsistencies in 'usertype' column
all_trips['usertype'].value_counts()
```

```python
#Data cleaning
#Checking if there are any rows corresponding to some kind of maintenance in 'from_station_name' column
all_trips['from_station_name'].value_counts()
```

```python
#Data formatting
#Converting 'start_time' & 'end_time' columns into datetime values
all_trips['start_time'] = pd.to_datetime(all_trips['start_time'])
all_trips['end_time'] = pd.to_datetime(all_trips['end_time'])
all_trips.info()
```

```python
#Data formatting
#Extracting date, day, month from 'start_time' column & droping 'tripduration' column as it's a string type
all_trips['date'] = all_trips['start_time'].dt.date
all_trips['day'] = all_trips['start_time'].dt.day
all_trips['month'] = all_trips['start_time'].dt.month
all_trips = all_trips.drop(columns = ['tripduration'])
all_trips.info()
```

```python
#Data formatting
#Creating a tripduration_min column
all_trips['tripduration'] = (all_trips['end_time'] - all_trips['start_time']).dt.total_seconds()
all_trips['tripduration_min'] = all_trips['tripduration'] / 60
all_trips.info()
```

```python
#Data cleaning
#Checking outlier values for tripduration_min
all_trips['tripduration_min'].describe()
```

```python
#Data cleaning
#Removing trip durations less than 1 min & more than 24h, 
#as well as rows from from_station_name indicating some kind of maintenance
all_trips_v2 = all_trips[
    (all_trips['from_station_name'] != "LBS - BBB La Magie") & 
    (all_trips['from_station_name'] != "Special Events") &
    (all_trips['from_station_name'] != "TS ~ DIVVY PARTS TESTING") &
    (all_trips['from_station_name'] != "BBB ~ Divvy Parts Testing") &
    (all_trips['tripduration_min'] >= 1) & 
    (all_trips['tripduration_min'] <= 1440)].copy()

all_trips_v2['tripduration_min'].describe()
```

```python
#Data cleaning
#Checking cleaning in 'from_station_name' column
all_trips_v2['from_station_name'].value_counts()
```

```python
#Data formatting
#Transforming 'date' column to datetime type & then extracting day_of_week
all_trips_v2['date'] = pd.to_datetime(all_trips_v2['date'])
all_trips_v2['day_of_week'] = all_trips_v2['date'].dt.day_name()
all_trips_v2.head()
```

```python
#Data cleaning
#Checking 'birthyear' inaccuracies 
all_trips_v2['birthyear'].describe()
```

```python
#Data cleaning
#Checking how many rows are probably inaccurate for 'birthyear' column 
trip_year = 2019
age = trip_year - all_trips_v2['birthyear']

suspect_rows = all_trips_v2[(age > 90) | (age < 10)]
suspect_rows.shape[0]
```

```python
#Data cleaning
#Removing 'birthyear' probably inaccurate rows
all_trips_v3 = all_trips_v2[
    (age < 90) & (age > 10)].copy()
all_trips_v3['birthyear'].describe()
```

```python
#Data analysis
#User type distribution
all_trips_v3['usertype'].value_counts()
```

```python
#Data analysis
#Gender distribution
all_trips_v3['gender'].value_counts()
```

```python
#Data analysis
#Usertype distribution
all_trips_v3['usertype'].value_counts()
```

```python
#Data analysis
#Age distribution
all_trips_v3['age'] = trip_year - all_trips_v3['birthyear']

print("\nAge statistics grouped by gender & usertype:\n", 
    all_trips_v3.groupby(['gender', 'usertype'])['age'].agg(['count','mean', 'median', 'min', 'max']))
```

```python
#Data analysis
#Age groups (under or over 30yo)
print("\nNumber of users older than 30yo:\n", 
    all_trips_v3[(all_trips_v3['age'] > 30)].groupby(['gender', 'usertype'])['age'].agg(['count']))
```

```python
#Data analysis
#Age groups (under or over 30yo)
print("\nNumber of users younger than 30yo:\n", 
all_trips_v3[(all_trips_v3['age'] <= 30)].groupby(['gender', 'usertype'])['age'].agg(['count']))
```

```python
#Data analysis
#Age groups (under or over 30yo) & trip duration
print("\nTrip duration statistics of users older than 30yo:\n", 
    all_trips_v3[(all_trips_v3['age'] > 30)].groupby(['gender', 'usertype'])['tripduration_min'].agg(['mean', 'median']))
```

```python
#Data analysis
#Age groups (under or over 30yo) & trip duration
print("\nTrip duration statistics of users younger than 30yo:\n", 
    all_trips_v3[(all_trips_v3['age'] < 30)].groupby(['gender', 'usertype'])['tripduration_min'].agg(['mean', 'median']))
```

```python
#Data analysis
#Trip duration regarding user type
print("\nTrip duration statistics grouped by usertype:\n", 
      all_trips_v3.groupby('usertype')['tripduration_min'].agg(['mean', 'median']))
```

```python
#Data analysis
#Trip duration statistics regarding user type & day of the week
weekly_stats = all_trips_v3.groupby(['usertype', 'day_of_week'])['tripduration_min'].agg(['count', 'mean', 'median'])
weekly_stats
```

```python
#Data analysis
#Trip duration statistics regarding user type & month of the year
monthly_stats = all_trips_v3.groupby(['usertype', 'month'])['tripduration_min'].agg(['count', 'mean', 'median'])
monthly_stats
```

```python
#Data analysis
#Calculating percentage of roundtrips depending on usertype
all_trips_v3['round_trip'] = (
    all_trips_v3['from_station_name']
    == all_trips_v3['to_station_name']
)

all_trips_v3.groupby('usertype')['round_trip'].mean() * 100
```

```python
#Data analysis
#Calculating average tripduration on roundtrips depending on usertype
print("\nAverage trip duration on roundtrips by user type:\n", 
    all_trips_v3.groupby(['usertype', 'round_trip'])['tripduration_min'].mean())
```

```python
#Data analysis
#Calculating top 5 pair stations depending on usertype
top_pairs = (all_trips_v3.groupby([
        'usertype',
        'from_station_name',
        'to_station_name'
    ])
    .size()
    .reset_index(name='number_of_rides')
    .sort_values('number_of_rides', ascending=False)
)

topstations_subs = top_pairs[top_pairs['usertype'] == 'Subscriber'].head(5)
topstations_subs
```

```python
#Data analysis
#Calculating top 5 pair stations depending on usertype
topstations_cust = top_pairs[top_pairs['usertype'] == 'Customer'].head(5)
topstations_cust
```

```python
#Saving all stats to create visualizations
weekly_stats.to_csv('bikeshare2018to2019_weeklystats.csv')
monthly_stats.to_csv('bikeshare2018to2019_monthlystats.csv')
topstations_subs.to_csv('bikeshare2018to2019_topstations_subs.csv', index=False)
topstations_cust.to_csv('bikeshare2018to2019_topstations_cust.csv', index=False)
```
